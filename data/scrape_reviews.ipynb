```python
from google_play_scraper import reviews_all, Sort
import csv

# Scraping ulasan Roblox
scrapreview = reviews_all(
    'com.roblox.client',
    lang='id',
    country='id',
    sort=Sort.MOST_RELEVANT,
    count=10000
)

# Simpan ke CSV
with open('roblox_reviews.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Review', 'Score'])
    for review in scrapreview:
        writer.writerow([review['content'], review['score']])

print(f"Scraping selesai, disimpan ke roblox_reviews.csv")
```
